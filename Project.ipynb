Now can the code columnm be set as the index. In order to be so it must be unique:

```python
df.code.is_unique
```

The codes are not unique so let's check what those values are:

```python
duplicate_codes_indices = np.where(df.duplicated('code', keep=False).values)[0]
dups = df.iloc[duplicate_codes_indices].copy()
dups.code = dups.code.astype(float)
dups.sort_values(by = 'code', inplace = True)
dups
```

As it can be seen there are two types of duplicates:
Some codes appear twice and the some code are NaN. As it can be observed when code are duplicated they are often the same and sometimes one row contain more information than the other. So we could just keep the rows that contain the most information. And for the rows with NaN as code, the values of the columns are not matching the column so there must be a shift of values. As a lot of values are missing, it would be complex to recover the true data and there are only tens of code with NaNs, we can delete them:

```python
dups = dups[dups != float('nan')]

dups['nb_of_features'] = np.sum(dups.notnull(), axis=1)
dups.reset_index(inplace = True)
indices_to_remove = dups.sort_values(['code','nb_of_features'],ascending = True)\
                        .groupby('code',as_index=False).first()['index'].values
df.dropna(subset=['code'],inplace=True)
df.drop(indices_to_remove,inplace = True)
```

### Removing useless columns

In order to answer our research questions, we do not need all the columns such as *creator* or *created_datetime* among others. We decided to drop all the columns that are not relevant to our research questions. This is done manually by looking at the columns in the dataframe.

```python
columns_to_drop = ['url',
                  'creator',
                  'created_t',
                  'created_datetime',
                  'last_modified_t',
                  'last_modified_datetime',
                  'generic_name',
                  'quantity',
                  'countries_tags',
                  'brands_tags',
                  'origins',
                  'countries',
                  'origins_tags',
                  'emb_codes',
                  'emb_codes_tags',
                  'first_packaging_code_geo',
                  'cities',
                  'cities_tags',
                  'purchase_places',
                  'stores',
                  'traces',
                  'traces_tags',
                  'serving_size',
                  'no_nutriments',
                  'additives',
                  'additives_tags',
                  'image_url',
                  'allergens', 
                  'image_small_url',
                  'ph_100g',
                  'manufacturing_places_tags',
                  'pnns_groups_1',
                  'pnns_groups_2',
                  'image_ingredients_url',
                  'image_ingredients_small_url',
                  'image_nutrition_url',
                  'image_nutrition_small_url',
                  'states',        
                  'states_en',
                  'states_tags',
                  'serving_quantity',
                  'traces_en',
                  'allergens_en',
                  'categories',
                  'categories_tags',
                  'packaging_tags',
                  'additives_en',
                  'ingredients_that_may_be_from_palm_oil',
                  'ingredients_from_palm_oil'
                  ]
```

Some columns that are deleted are obvious not part of our plans (for instance *image_nutrition_url* or *first_packaging_code_geo*) while others required more in-depth analysis. Indeed, a large number of columns are divided into 3 columns with redundancy. For example, the *countries* column also has *countries_en* for the english name and *countries_tags* for the tag representing that country. By analyzing the three columns, we found that *countries_en* has the clearest names (while others have different names for the same country). Hence, we decided to only keep the *countries_en* column. The same analysis has been done for all the columns that present the same situation. Finally, a few columns such as *ingredients_from_palm_oil* only have NaN values and are thus removed from the dataframe.

```python
# We drop all the columns defined above
df_new = df.drop(columns_to_drop, axis=1)
df_new
```

This current data-set still has 124 columns, all of them that can be used for our research questions. Currently, the columns are dominated by the ingredients. However, there is a possibility that some ingredients have mostly *NaN* values, thus would not help our analysis and made the data not clean. As a result, we decided to analyze the data for each ingredient i.e. the number of values that are not *NaN* for every ingredients.

### Analyzing the ingredients

Most of the columns at the end represent ingredients that are important for our research questions. However, it seems that most of the ingredients contain too much *NaN* values. We would like to check, for each ingredient, how many non-NaN values it has.

```python
# We take the column indices that are ingredients
l1 = list(range(20, 114))
l2 = list(range(118, df_new.shape[1]))

# We take the columns that represent ingredients to analyze them
df_ingredients = df_new[df_new.columns[l1 + l2]]

# For each ingredient, we see how much times it appears (how many non-NaN values)
ingredients_count = df_ingredients.count().sort_values(ascending=False)

# We see the 30 ingredients that have the less NaN values
ingredients_count[0:30]
```

We notice that some ingredients only have values for 1 product out of 699098. In fact, most of the ingredients do not have values for more than 4000 products out of 699098 in total. These columns can then be dropped because they are not interesting values since they do not add too much value. Taking the top 30 was arbitrarily chosen because 30 is around 1/3 of the total number of ingredients which is good enough for our analysis

```python
# We convert the above series into a dataframe to easily take the names of the ingredients and put them
# in a list. Note that we take the ingredients that have less than 4000 non-NaN values.
ingredients_count_df = pd.DataFrame(ingredients_count).reset_index()[30:94]

# We take the ingredients that are not in top 30 and put them in a list
ingredients_to_drop = ingredients_count_df['index'].tolist()

# We drop all those ingredients
df_no_ingredients = df_new.drop(ingredients_to_drop, axis=1)
df_no_ingredients.head(10)
```

This new dataframe has 60 columns, which is an improvement form the original 169 columns. More than 100 columns were either useless for our research questions or do not have enough data to be interesting.
