# Data cleaning

## Loading the CSV file

```python
path = ""
file_name = "OpenFood.csv"
data_path = path+file_name
df = pd.read_csv(data_path, sep = '\t')
```

### DataFrame properties

Are there some duplicated rows?

```python
np.sum(df.duplicated().values == True)
```

So the duplicated values are deleted:

```python
df.drop_duplicates(inplace = True)
```

We can also drop the column that only conain NaN values:

```python
df.dropna(axis = 1, how='all', inplace = True)
```

```python
df.shape
```

So now there are 683257 rows and 196 columns

Now can the code columnm be set as the index. In order to be so it must be unique:

```python
df.code.is_unique
```

The codes are not unique so let's check what are those values are:

```python
duplicate_codes = np.where(df.duplicated('code', keep=False).values)[0]
dups = df.iloc[duplicate_codes]
dups.code = dups.code.astype(float)
dups.sort_values(by = 'code')
```

As it can be seen there are two types of duplicates:
Some codes appear twice and the some code are NaN. As it can be observed when code are duplicated they are often the same and sometimes one row contain more information than the other. So we could just keep the rows that contain the most information. And for the rows with NaN the values of the columns are not matching the column nae so there must be a shift of values. As a lot of values are missing, it would be complex to recover the true data and there are only tens of code with NaNs, we can delete them:

```python
df = df[df.code.astype(float) != float('nan')]
```
